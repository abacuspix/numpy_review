```python
import numpy as np
import timeit
import matplotlib.pyplot as plt

# This program measures the performance of the NumPy sort function
# and plots time vs array size.

def dosort():
    integers.sort()

def measure():
    timer = timeit.Timer('dosort()', 'from __main__ import dosort, integers')
    return timer.timeit(10 ** 2)

# Generate sizes as powers of 2 from 2^0 to 2^18
powers_of_2 = np.arange(0, 19)
sizes = 2 ** powers_of_2

times = []

for size in sizes:
    integers = np.random.randint(1, 10 ** 6, size)
    times.append(measure())

times = np.array(times)

# Fit a polynomial (log-log space)
log_sizes = np.log2(sizes)
log_times = np.log2(times)
fit = np.polyfit(log_sizes, log_times, 1)

# Plotting
plt.figure(figsize=(10, 6))
plt.title("Sort array sizes vs execution times")
plt.xlabel("Array Size (log scale)")
plt.ylabel("Execution Time (s) (log scale)")
plt.loglog(sizes, times, 'ro', label='Measured times')
plt.loglog(sizes, 2 ** np.polyval(fit, log_sizes), label=f'Fit: y = {2 ** fit[1]:.2e} * x^{fit[0]:.2f}')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import numpy as np
import unittest

def factorial(n):
    if n == 0:
        return np.array([1])

    if n < 0:
        raise ValueError("Unexpected negative value")

    return np.arange(1, n + 1).cumprod()

class FactorialTest(unittest.TestCase):
    def test_factorial(self):
        # Test for the factorial of 3 that should pass.
        self.assertEqual(6, factorial(3)[-1])
        np.testing.assert_equal(np.array([1, 2, 6]), factorial(3))

    def test_zero(self):
        # Test for the factorial of 0 that should pass.
        self.assertEqual(1, factorial(0)[0])

    def test_negative(self):
        # Test for the factorial of negative numbers that should raise a ValueError.
        self.assertRaises(ValueError, factorial, -10)

# If running interactively, we can use the following:
if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)
```

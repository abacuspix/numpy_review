```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

# Calculate the future value for the given parameters
print("Future value", np.fv(0.03/4, 5 * 4, -10, -1000))

# List to store future values
fvals = []

# Calculate future values for periods from 1 to 9 years
for i in range(1, 10):
    fvals.append(np.fv(0.03/4, i * 4, -10, -1000))

# Plot the future values
plt.plot(range(1, 10), fvals, 'bo')
plt.title('Future value, 3 % interest,\n Quarterly payment of 10')
plt.xlabel('Saving periods in years')
plt.ylabel('Future value')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
from __future__ import print_function
import numpy as np
import datetime

def datestr2num(s):
    return datetime.datetime.strptime(s.decode('utf-8'), "%d-%m-%Y").toordinal()

# Load dates and closing prices from the CSV file
dates, closes = np.loadtxt('AAPL.csv', delimiter=',', usecols=(1, 6), converters={1: datestr2num}, unpack=True)

# Sort indices based on dates and closing prices
indices = np.lexsort((closes, dates))

print("Indices", indices)

# Print the sorted dates and closing prices
sorted_dates_closes = ["%s %s" % (datetime.date.fromordinal(int(dates[i])), closes[i]) for i in indices]
print(sorted_dates_closes)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import datestr2num

# Load closing prices from 'AAPL.csv'
closes = np.loadtxt('AAPL.csv', delimiter=',', usecols=(6,), converters={1: datestr2num}, unpack=True)

# Define the Blackman window length
N = 5

# Generate the Blackman window
window = np.blackman(N)

# Apply the Blackman window to smooth the data
smoothed = np.convolve(window / window.sum(), closes, mode='same')

# Plot the smoothed data and original closing prices
plt.plot(smoothed[N:-N], lw=2, label="Smoothed")
plt.plot(closes[N:-N], '--', label="Closes")
plt.title('Blackman Window Smoothing')
plt.xlabel('Days')
plt.ylabel('Price ($)')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of values from 0 to 4 with 100 points
x = np.linspace(0, 4, 100)

# Create a 2D grid by computing the outer product of x with itself
xx = np.outer(x, x)

# Compute the sinc function on the 2D grid
vals = np.sinc(xx)

# Display the 2D sinc function
plt.imshow(vals, extent=(0, 4, 0, 4), origin='lower', cmap='viridis')
plt.title('2D Sinc Function')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='sinc(xy)')
plt.grid(False)  # Grid lines may not be necessary for this kind of plot
plt.show()
```

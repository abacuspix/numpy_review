```python
import numpy as np
from matplotlib import pyplot as plt

# Number of trials
N = 10000

# Generate geometric outcomes with success probability of 1/6
outcome = np.random.geometric(1.0/6.0, size=N)

# Initialize enemy sequence
enemy = np.zeros(N) 
enemy[0] = outcome[0]

# Generate the cumulative sum of outcomes for the enemy sequence
for i in range(1, len(outcome)):
   enemy[i] = outcome[i] + enemy[i - 1]

# Generate the 'us' sequence as multiples of 6
us = 6 * np.arange(1, len(enemy) + 1)

# Generate time points for plotting
t = np.arange(N)

# Plot the enemy sequence
plt.plot(t, enemy, lw=1, label='Enemy')

# Plot the us sequence
plt.plot(t, us, lw=2, label='Us')

# Add title and labels
plt.title('Geometric Distribution vs Linear Progression')
plt.xlabel('Time')
plt.ylabel('Cumulative Value')
plt.legend()

# Show the plot
plt.show()
```

```python
from __future__ import print_function
import numpy as np

# Define matrix A
A = np.mat("3 -2; 1 0")
print("A\n", A)

# Calculate eigenvalues of matrix A
eigenvalues = np.linalg.eigvals(A)
print("Eigenvalues", eigenvalues)

# Calculate eigenvalues and eigenvectors of matrix A
eigenvalues, eigenvectors = np.linalg.eig(A)
print("First tuple of eig", eigenvalues)
print("Second tuple of eig\n", eigenvectors)

# Verify the eigenvalue equation A * v = λ * v
for i, eigenvalue in enumerate(eigenvalues):
      print("Left", np.dot(A, eigenvectors[:, i]))
      print("Right", eigenvalue * eigenvectors[:, i])
      print()
```

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

# Generate a cosine wave
x = np.linspace(0, 2 * np.pi, 30)
wave = np.cos(x)

# Perform Fourier transform
transformed = np.fft.fft(wave)

# Verify that the inverse Fourier transform yields the original wave
print(np.all(np.abs(np.fft.ifft(transformed) - wave) < 10 ** -9))

# Plot the magnitude of the transformed data
plt.plot(np.abs(transformed))
plt.title('Transformed cosine')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a cosine wave
x = np.linspace(0, 2 * np.pi, 30)
wave = np.cos(x)

# Perform Fourier transform
transformed = np.fft.fft(wave)

# Shift zero-frequency component to the center
shifted = np.fft.fftshift(transformed)

# Verify that the inverse shift yields the original transformed data
print(np.all(np.abs(np.fft.ifftshift(shifted) - transformed) < 10 ** -9))

# Plot the real parts of the transformed and shifted data
plt.plot(np.real(transformed), lw=2, label="Transformed (Real)")
plt.plot(np.real(shifted), '--', lw=3, label="Shifted (Real)")

# Plot the imaginary parts of the transformed and shifted data
plt.plot(np.imag(transformed), lw=2, label="Transformed (Imag)", alpha=0.6)
plt.plot(np.imag(shifted), '--', lw=3, label="Shifted (Imag)", alpha=0.6)

plt.title('Shifted and transformed cosine wave')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

# Initialize cash array and set starting cash
cash = np.zeros(10000)
cash[0] = 1000

# Set random seed for reproducibility
np.random.seed(73)

# Generate outcomes using binomial distribution
outcome = np.random.binomial(9, 0.5, size=len(cash))

# Simulate betting process
for i in range(1, len(cash)):
    if outcome[i] < 5:
        cash[i] = cash[i - 1] - 1
    elif outcome[i] < 10:
        cash[i] = cash[i - 1] + 1
    else:
        raise AssertionError("Unexpected outcome " + str(outcome[i]))

# Print minimum and maximum outcome for verification
print("Minimum outcome:", outcome.min(), "Maximum outcome:", outcome.max())

# Plot the results
plt.plot(np.arange(len(cash)), cash)
plt.title('Binomial Simulation of Betting Process')
plt.xlabel('Number of Bets')
plt.ylabel('Cash')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of samples
N = 10000
# Set the random seed for reproducibility
np.random.seed(34)

# Generate lognormal distributed values
lognormal_values = np.random.lognormal(size=N)

# Plot the histogram
_, bins, _ = plt.hist(lognormal_values, bins=int(np.sqrt(N)), density=True, lw=1, label="Histogram")

# Define sigma and mu for the lognormal distribution
sigma = 1
mu = 0

# Calculate the PDF
x = np.linspace(min(bins), max(bins), len(bins))
pdf = np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi))

# Plot the PDF
plt.xlim([0, 15])
plt.plot(x, pdf, '--', lw=3, label="PDF")

# Add titles and labels
plt.title('Lognormal distribution')
plt.xlabel('Value')
plt.ylabel('Normalized frequency')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of samples
N = 10000

# Set the random seed for reproducibility
np.random.seed(27)

# Generate normal distributed values
normal_values = np.random.normal(size=N)

# Plot the histogram
_, bins, _ = plt.hist(normal_values, bins=int(np.sqrt(N)), density=True, lw=1, label="Histogram")

# Define sigma and mu for the normal distribution
sigma = 1
mu = 0

# Calculate the PDF
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2))

# Plot the PDF
plt.plot(bins, pdf, '--', lw=3, label="PDF")

# Add titles and labels
plt.title('Normal distribution')
plt.xlabel('Value')
plt.ylabel('Normalized Frequency')
plt.grid()
plt.legend(loc='best')
plt.show()
```
